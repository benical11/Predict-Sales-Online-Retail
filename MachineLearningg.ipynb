```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

sns.set()
```

```python
df = pd.read_csv("data.csv").drop(["Unnamed: 0"], axis=1)
df.tail(1000)
```

### InvoiceDate'i index olarak düzenleyip int ve float olmayan verileri categorical olarak çevirdik. Bunun nedeni veriyi öğrenirken farklı bir veri girişi ile hata almamak için

```python
df2 = df.set_index("InvoiceDate")
```

```python
df2 = pd.get_dummies(df2, drop_first = True)
df2.head()
```

### Quantity sütunumuz tahmin etmek istediğimiz verimiz olduğu için X'e target verimiz dışındakileri attık y'ye ise sadece Quantity verimizi attık

```python
X = df2.drop("Quantity", axis=1)
y = df2["Quantity"]

# Verimizi test_size'ını 0.2 olarak belirleyerek 80 e 20'lik bir split(ayırma) işlemi gerçekleştirdik 
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
X_train.head()
```

### Linear Regression ile modelin eğitimini yaptık

```python
reg = LinearRegression()

reg.fit(X_train, y_train)
```

### Elimizde olan verilerle %70'lik bir score'a ulaştık 

```python
reg.score(X_test, y_test)
```

### Tahmin verilerini alabilmek için modelin test verisinden aldığı predict fonksiyonu ile tahmin listesi oluşturduk

```python
y_pred = reg.predict(X_test)
y_pred
```

### Görülebilmesi için de harcama ve tahmin olmak üzere iki sütunlu bir dataframe oluşturum bunları içine attık

```python
def undummify(df, prefix_sep="_"):
    cols2collapse = {
        item.split(prefix_sep)[0]: (prefix_sep in item) for item in df.columns
    }
    series_list = []
    for col, needs_to_collapse in cols2collapse.items():
        if needs_to_collapse:
            undummified = (
                df.filter(like=col)
                .idxmax(axis=1)
                .apply(lambda x: x.split(prefix_sep, maxsplit=1)[1])
                .rename(col)
            )
            series_list.append(undummified)
        else:
            series_list.append(df[col])
    undummified_df = pd.concat(series_list, axis=1)
    return undummified_df
```

```python
df3 = undummify(X_test)
```

```python
pred_df = pd.DataFrame(columns=["Satış Miktarı","tahmin"], index=X_test.index)

pred_df = df3
pred_df["Satış Miktarı"] = y_test
pred_df["tahmin"] = y_pred

pred_df.tail(10)
```
